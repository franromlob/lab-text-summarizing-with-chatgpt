# Text summarizing with ChaptGPT
In this lesson, you will summarize text with a focus on specific topics.

## Setup

````python
from openai import OpenAI
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file

OPENAI_API_KEY  = os.getenv('OPENAI_API_KEY')
````

````python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)


def get_completion(prompt, model="gpt-3.5-turbo"): # Andrew mentioned that the prompt/ completion paradigm is preferable for this class
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message.content
````

## Text to summarize

````python
prod_review = """
Got this panda plush toy for my daughter's birthday, \
who loves it and takes it everywhere. It's soft and \ 
super cute, and its face has a friendly look. It's \ 
a bit small for what I paid though. I think there \ 
might be other options that are bigger for the \ 
same price. It arrived a day earlier than expected, \ 
so I got to play with it myself before I gave it \ 
to her.
"""
````

## Summarize with a word/sentence/character limit

````python
prompt = f"""
Your task is to generate a short summary of a product \
review from an ecommerce site. 

Summarize the review below, delimited by triple 
backticks, in at most 30 words. 

Review: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````

# 📊 Technical Report: Automated Linguistic Analysis via Zero-Shot Prompting

## 1. Project Overview
**Subject:** C1 English Structure & Vocabulary Extraction  
**Objective:** To automate the identification of advanced grammatical structures and high-level vocabulary from three distinct text types (Futuristic Cities, Consumption, and Sports).

---

## 2. Experimental Findings (Report)

In this experiment, I used a **Zero-Shot** approach within a Python loop. While the model successfully identified the linguistic components, the following structural failures were observed:

| Variation | Behavior | Result |
| :--- | :--- | :--- |
| **Iteration 0 (Text 0)** | Followed the provided template correctly. | **Success.** |
| **Iteration 1 (Text 1)** | **Naming Inconsistency.** It changed the title to "Environmental Impact..." instead of using the provided titles. | **Partial Failure.** |
| **Iteration 2 (Text 1)** | **Indexing Error & Hallucination.** It labeled the third text as "Text 1" again and used vocabulary from previous texts (`daunting`, `mitigate`) rather than the actual text provided. | **Failure (Hallucination).** |

### **Hallucinations & Logic Failures:**
* **Contextual Drift:** In the second and third responses, GPT started mixing vocabulary from the example prompt (`mitigate`, `veritable`, `ubiquitous`) with the actual text. This is a clear sign that the model was **over-influenced by the example** and failed to extract data solely from the source text.
* **Index Failure:** The loop index (`i`) was ignored or overwritten by the model's internal prediction, resulting in "Text 1" being repeated twice.



---

## 3. What I Learned (Technical Insights)

### 🧠 Strategic Takeaways:

* **The Limitation of Zero-Shot in Loops:** I learned that when using a loop in Python to call an LLM, providing a single example within the string (Zero-Shot with a template) often leads to **Pattern Over-fitting**. The model begins to copy the example data instead of analyzing the new input.
* **Output Determinism:** Without strict formatting (like **Few-Shot** or **JSON Schema**), the model's output becomes unpredictable. For an **AI Engineer**, this is a critical failure because the data cannot be easily parsed into a database or a CSV.
* **The Need for System Roles:** Using only a `user` prompt for complex extraction often causes the model to "lose its way." Defining a clear `system` role (e.g., "You are a linguistic analyzer") would help maintain focus on the specific text provided.
* **Prompt Leakage:** The "Example" I provided in the prompt was so strong that the model "leaked" that example vocabulary into the final results, even when those words didn't exist in the input text.



---

## 4. Professional Conclusion
This exercise proves that for **Solutions Architecture**, we cannot rely on a single string template for repetitive tasks. To fix this, I would implement a **Few-Shot** approach with at least two diverse examples and enforce a strict **JSON response format** to ensure that `Text {i}` remains accurate and that the vocabulary is exclusively pulled from the `texts` list.
